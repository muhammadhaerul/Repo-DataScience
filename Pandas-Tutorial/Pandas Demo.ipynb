```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
print(str(round(1/3*100, 3)) + '%')
```

```python
def percent(x):
    val = x.str.contains('Python').sum()
    total = x.count()
    return (str(round(val/total*100, 3)) + '%')
```

### 1. How many percent people from each countries known python?

```python
# first way
country_group['LanguageWorkedWith'].apply(percent)
```

```python
country_group['LanguageWorkedWith'].apply(percent).loc['Indonesia']
```

```python
country_group.get_group('Indonesia')['LanguageWorkedWith'].str.contains('Python').value_counts(normalize=True)*100
```

```python
# second way
country_respondent = df['Country'].value_counts()
country_respondent
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# concat as a row 
pd.concat([country_respondent, country_uses_python])
```

```python
# concat as column
pd.concat([country_respondent, country_uses_python], axis='columns')
```

```python
df2 = pd.concat([country_respondent, country_uses_python], axis='columns', sort=True)
```

```python
df2.rename(columns={'Country': 'Respondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
df2
```

```python
df2['PercentageKnowsPython'] = df2['NumKnowsPython'] / df2['Respondents'] * 100
```

```python
df2 = df2.sort_values(by='PercentageKnowsPython', ascending=False)
df2.head(20)
```
