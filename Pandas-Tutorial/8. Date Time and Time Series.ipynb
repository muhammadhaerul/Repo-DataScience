## Reread our Datasets

```python
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p') 
```

```python
df = pd.read_csv('data/ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df
```

```python
df.loc[0, 'Date']
```

### Convert datetime to day name

```python
df.loc[0, 'Date'].day_name()
```

### Filtering Date

```python
filt = (df['Date'] > '2020')
filt
```

```python
filt1 = (df['Date'] >= '2020')
df.loc[filt1]
```

```python
filt2 = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt2]
```

```python
filt3 = (df['Date'] >= pd.to_datetime('2019-01-01'))
df.loc[filt3]
```
