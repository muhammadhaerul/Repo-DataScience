```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/people2.csv')
df
```

```python
people = {
    'name': ['Nou', 'Tamuz', np.nan],
    'clan': ['Polaris', np.nan, None],
    'email': ['None', 'tamus@bulan.com', 'Missing'],
    'age': [15, 'NA', None]
}

df2 = pd.DataFrame(people)
df2
```

```python
df3 = pd.concat([df, df2], ignore_index=True)
df3
```

```python
df3['age'] = [17, 16, 17, 20, 35, 30, 15, 100, None]
```

```python
df = df3
df
```

## Droping Missing Values

### Dropping by index

```python
df.dropna()
```

```python
df.dropna(axis='index', how='any') #default
```

```python
df.dropna(axis='index', how='all') #drop only if all columns in a row is a missing value
```

### Dropping by columns

```python
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='columns', how='all')
```

### Dropping spesific values

```python
df.dropna(axis='index', how='any', subset=['clan'])
```

```python
df.dropna(axis='index', how='any', subset=['name', 'clan'])
```

```python
df.dropna(axis='index', how='all', subset=['name', 'clan'])
```

## Replace missing values

```python
people = {
    'name': ['Nou', 'Tamuz', np.nan],
    'clan': ['Polaris', np.nan, None],
    'email': ['None', 'tamus@bulan.com', 'Missing'],
    'age': [15, 'NA', None]
}

df4 = pd.DataFrame(people)
df4
```

```python
df5 = df4.replace('NA', np.nan)
df5.replace('Missing', np.nan, inplace=True)
df5
```

```python
df = pd.concat([df3, df5], ignore_index=True)
df
```

```python
df.dropna(axis='index', how='all', subset=['name', 'clan'])
```

## Handling Missing Values

```python
df.isna() #check valus whether is it missing or not?
```

```python
df.fillna('MISSING') #fill the missing values with a spesifiv value
```

```python
df['age'].fillna(0)
```

```python
print(df['age'].mean())
print(df['age'].median())
```

```python
age_ser = round(df['age'].fillna(df['age'].mean()))
age_ser
```

```python
df['age'] = age_ser
df
```

## Casting Datatype Series

```python
df.dtypes
```

```python
type(np.nan)
```

```python
df['age'] = df['age'].astype(int)
df
```

## Dropping Duplicate Rows

```python
df.replace('Missing', np.nan, inplace=True)
```

```python
# keep first duplicate row
df.drop_duplicates(inplace=True)
df
```

## Full Cleaning

```python
df.dropna(axis='index', how='any', subset=['name', 'clan'], inplace=True)
df
```

```python
update_email = (df['name'] + '@' + df['clan'] + '.com').str.replace(' ', '').str.lower()
update_email
```

```python
df['email'] = update_email
df
```

```python
df.to_csv('data/people3.csv', index=False)
```
