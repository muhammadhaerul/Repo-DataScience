```python
import pandas as pd
```

## CSV

```python
df = pd.read_csv('data/survey_sof_2022.csv', index_col='Respondent')
```

```python
pd.set_option('display.max_columns', df.shape[1])
```

```python
df
```

```python
idn_filt = (df['Country'] == 'Indonesia')
idn_df = df.loc[idn_filt]
idn_df
```

### Custome seperator

```python
idn_df.to_csv('data/idn_survey_sof_2022.csv') #by default the sep is a comma
```

```python
idn_df.to_csv('data/idn_survey_sof_2022.tsv', sep='\t')
```

## Databases

```python
from sqlalchemy import create_engine
import psycopg2
```

```python
postgres_str = (f"postgresql://"
                f"{'postgres'}:{'postgres'}@"
                f"{'localhost'}:{'5432'}/{'mydatabase'}")
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/mydatabase')
```

```python
engine = create_engine(postgres_str) 
```

```python
idn_df.to_sql('idn_table', engine) #write dataframe to postgres db
```

```python
idn_df.to_sql('idn_table', engine, if_exists='replace')
```

```python
sql_df = pd.read_sql('idn_table', engine, index_col='Respondent')
```

```python
sql_df
```

```python
sql_df = pd.read_sql_query('SELECT * FROM idn_table', engine, index_col='Respondent')
sql_df.head()
```
