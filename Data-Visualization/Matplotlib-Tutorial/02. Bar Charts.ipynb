```python
# read data from file
# csv module

import csv
from collections import Counter

with open('data/sof_survey.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    #row = next(csv_reader)
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_counter.most_common(15))
```

```python
# split the languange_counter

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# reverse the list
languages.reverse()
popularity.reverse()


print(languages)
print(popularity)
```

```python
# crate the bar charts
plt.style.use('fivethirtyeight')

# use barh not bar for horizontal layout
plt.barh(languages, popularity)

plt.title('Most Popular Languanges')
plt.ylabel('Programming Languanges')
plt.xlabel('Number of People Who Use')

plt.tight_layout()

plt.show()
```

```python
# pandas library
df = pd.read_csv('data/sof_survey.csv', index_col='Responder_id')

language_counter = Counter()

for response in df['LanguagesWorkedWith']:
    language_counter.update(response.split(';'))

languages, popularity = [], []

for r in language_counter.most_common(28):
    languages.append(r[0])
    popularity.append(r[1])

languages.reverse()
popularity.reverse()
```

```python
# crate the bar charts
plt.style.use('fivethirtyeight')

# use barh not bar for horizontal layout
plt.barh(languages, popularity)

plt.title('Most Popular Languanges')
plt.ylabel('Programming Languanges')
plt.xlabel('Number of People Who Use')

plt.tight_layout()

# save fig
plt.savefig('fig/bar-chart.png')

plt.show()
```
