```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

## Line Plots

```python
# read and load data
df = pd.read_csv('data/devs_by_age.csv')
df.head()
```

```python
# split dataframe to list
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']
```

```python
# median salaries
overall = dev_salaries.append(py_salaries).append(js_salaries)
overall_median = overall.median()
overall_median
```

```python
# create line plot
plt.plot(ages, dev_salaries, color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
#plt.plot(ages, js_salaries, label='JavaScript Devs')

# fill
plt.fill_between(ages, py_salaries, overall_median,
                 interpolate=True, alpha=0.25)

plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
# create line plot
plt.plot(ages, dev_salaries, color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
#plt.plot(ages, js_salaries, label='JavaScript Devs')

# fill between
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries < overall_median),
                 interpolate=True, alpha=0.25)

plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
# create line plot
plt.plot(ages, dev_salaries, label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
#plt.plot(ages, js_salaries, label='JavaScript Devs')

# fill between
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, color='green', alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries < dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

# save fig
plt.savefig('fig/line-plot.png')

plt.show()
```
