## Pie Charts

```python
"""
Colors:
Blue = #008fde5
Red = #fc4f30
Yellow = #e5ae37
Green = #6d904f
"""
```

```python
# set style
plt.style.use('fivethirtyeight')
```

**Note:** Using 'semicolon' (';') to hide details output

```python
# create a pie chart
slices = [60, 40, 15, 20]
labels = ['Sixty', 'Fourty', 'Fifteen', 'Twenty']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'});
```

```python
# create a dummy data
slices = [60, 40, 15, 20]
labels = ['Sixty', 'Fourty', 'Fifteen', 'Twenty']

# using hexadecimal for coloring
colors = ['#fc4f30', '#e5ae37', '#6d904f', '#008fde5']

# create a pie chart
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

# set plot attribute
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# using real data

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels)

plt.title('Language Popularity')
plt.tight_layout()
plt.show()
```

```python
# Pie chart is not a good option to vizualise large data 
# Recommendation: 3-5 data 
```

```python
# Top 5 most popular languages
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# to explode out slices of pie
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode)

plt.title('Language Popularity')
plt.tight_layout()
plt.show()
```

```python
# Top 5 most popular languages
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# to explode out slices of pie
explode = [0, 0, 0, 0.1, 0]

# add more args
plt.pie(slices, labels=labels, explode=explode, 
        shadow=True, startangle=90, autopct='%1.1f%%')

plt.title('Language Popularity')
plt.tight_layout()

# save fig
plt.savefig('fig/pie-chart.png')

plt.show()
```
