```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.__version__
```

## Scatter Plots

```python
# create dummy data
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
# set style
plt.style.use('seaborn')
```

```python
# create scatter plots
plt.scatter(x, y)

plt.tight_layout()
plt.show()
```

```python
# create scatter plots
plt.scatter(x, y, s=50, c='green', edgecolor='black', 
            linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
```

```python
# create some dummy data 
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, edgecolor='black', cmap='Greens', 
            linewidth=1, alpha=0.75)

# add color bar
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.title('Scatter Plot')

plt.tight_layout()
plt.show()
```

```python
# using real data
df = pd.read_csv('data/viewers_respons.csv')
df
```

```python
# split dataframe to list
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
# create scatter plot
plt.scatter(view_count, likes, c=ratio, cmap='summer', 
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# set scale to log for avoiding big distance amount of data and outliers
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

# save fig
plt.savefig('fig/scatter-plot.png')

plt.tight_layout()
plt.show()
```
