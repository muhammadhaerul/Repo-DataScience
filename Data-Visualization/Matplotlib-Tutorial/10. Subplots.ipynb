```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

## Subplots

```python
# set style
plt.style.use('seaborn')
```

```python
# read and load data
df = pd.read_csv('data/devs_by_age.csv')
df.head()
```

```python
# split dataframe to list
ages = df['Age']
dev_sal = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']

# create plots
plt.plot(ages, dev_sal, label='All Devs')
plt.plot(ages, py_sal, label='Python Devs')
plt.plot(ages, js_sal, label='JavaScript Devs')

# add atributes
plt.legend()
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# split dataframe to list
ages = df['Age']
dev_sal = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']

# create sub plots
fig, ax = plt.subplots() #by default it's just 1 plot

# create plots
ax.plot(ages, dev_sal, label='All Devs')
ax.plot(ages, py_sal, label='Python Devs')
ax.plot(ages, js_sal, label='JavaScript Devs')

# add legend
ax.legend()

# add atributes
ax.set_title('Median Salary by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# split dataframe to list
ages = df['Age']
dev_sal = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']

# create sub plots (1 fig, multi ax)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), sharex=True) #by default it's just 1 plot

# create plots
ax1.plot(ages, dev_sal, label='All Devs')
ax2.plot(ages, py_sal, label='Python Devs')
ax2.plot(ages, js_sal, label='JavaScript Devs')

# add atributes for ax1
ax1.legend()
ax1.set_title('Devs Salary')
ax1.set_ylabel('Median Salary (USD)')

# add atributes ax2
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

# save fig
fig.savefig('fig/subplots.png')
```

```python
# split dataframe to list
ages = df['Age']
dev_sal = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']

# create sub plots (1 fig, 1 ax)
fig1, ax1 = plt.subplots(figsize=(8, 8), sharex=True)
fig2, ax2 = plt.subplots(figsize=(8, 8), sharex=True)

# create plots
ax1.plot(ages, dev_sal, label='All Devs')
ax2.plot(ages, py_sal, label='Python Devs')
ax2.plot(ages, js_sal, label='JavaScript Devs')

# add atributes for ax1
ax1.legend()
ax1.set_title('Devs Salary')
ax1.set_ylabel('Median Salary (USD)')

# add atributes ax2
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

# save fig
fig1.savefig('fig/subfig1.png')
fig2.savefig('fig/subfig2.png')
```
